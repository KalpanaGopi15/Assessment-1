```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv('insurance_pre.csv')
```

```python
data1
```

```python
data1=pd.get_dummies(data1,drop_first=True,dtype=int)
```

```python
data1
```

```python
indep=data1[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dep=data1['charges']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
para_gride={'criterion':['squared_error', 'absolute_error','friedman_mse', 'poisson'], 'max_features': ['auto','sqrt','log2'],'n_estimators':[10,100]}
```

```python
gride=GridSearchCV(RandomForestRegressor(),para_gride,refit=True,verbose=3,n_jobs=-1)
```

```python
gride.fit(x_train,y_train)
```

```python
re=gride.cv_results_
print (re)
```

```python
grid_predictions = gride.predict(x_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,grid_predictions)
print("The R_score value for best parameter {}:".format(gride.best_params_),r_score)
```
